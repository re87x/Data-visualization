### Question 2: Use Webscraping to Extract Tesla Revenue Data  

Using webscraping, we extracted Tesla’s quarterly revenue data and stored it in a DataFrame called `tesla_revenue`.  
We then displayed the last five rows using the **tail()** function.  

```python
# Display the last fivme rows of Tesla revenue data
print(tesla_revenue.tail(5))

````python

# الرابط
url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"

# إضافة User-Agent عشان يفتكر إنك متصفح
headers = {"User-Agent": "Mozilla/5.0"}

# تحميل الصفحة
html_data = requests.get(url, headers=headers).text

# استخراج الجداول
tables = pd.read_html(html_data)

print("عدد الجداول:", len(tables))

# نختار الجدول الثاني (فيه الإيرادات)
tesla_revenue = tables[1]

# إعادة تسمية الأعمدة
tesla_revenue.columns = ["Date", "Revenue"]

# تنظيف البيانات
tesla_revenue.dropna(inplace=True)
tesla_revenue["Revenue"] = tesla_revenue["Revenue"].str.replace(r',|\$', '', regex=True)
tesla_revenue["Revenue"] = pd.to_numeric(tesla_revenue["Revenue"], errors="coerce")

# عرض آخر 5 صفوف
print(tesla_revenue.tail(5).to_string())
````

### Question 4: Use Webscraping to Extract GME Revenue Data  

Display the last five rows of the `gme_revenue` dataframe using the `tail` function.

````python
# Statista: GameStop revenue
url = "https://www.statista.com/statistics/272889/revenue-of-gamestop/"

# Fetch HTML with a browser-like header
headers = {"User-Agent": "Mozilla/5.0"}
html = requests.get(url, headers=headers).text

# Parse all tables from the HTML
tables = pd.read_html(StringIO(html))
print("Tables found:", len(tables))  # optional

# Pick the first table (contains revenue)
gme_revenue = tables[0].copy()

# Normalize column names to: Date, Revenue
gme_revenue.columns = ["Date", "Revenue"]

# Clean and convert Revenue to numeric
gme_revenue["Revenue"] = (
    gme_revenue["Revenue"].astype(str).str.replace(",", "").str.strip()
)
gme_revenue["Revenue"] = pd.to_numeric(gme_revenue["Revenue"], errors="coerce")

# Drop empty rows and show last 5
gme_revenue = gme_revenue.dropna(subset=["Revenue"]).reset_index(drop=True)
gme_revenue.tail(5)
````

### Question 5: Plot Tesla Stock Graph
make_graph(tesla_data, "Tesla Stock Price Over Time")

````python
# Download Tesla stock data
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="1mo")   # last 1 month data

# Define the graph function
def make_graph(data, title):
    plt.figure(figsize=(10,5))
    plt.plot(data.index, data['Close'], color='blue')
    plt.title(title)
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.grid(True)
    plt.show()

# Call the function to plot Tesla stock data
make_graph(tesla_data, "Tesla Stock Price Over Time")
````

## Exercise 6: Plot Tesla Revenue Graph

We now plot Tesla's **historical revenue data** to visualize the company’s growth over time.  
The function is similar to the stock graph, but instead of stock prices, we use the revenue dataset.

````python
# Download Tesla stock data (last 5 years)
tesla = yf.Ticker("TSLA")
tesla_data = tesla.history(period="5y")

# Define the function to create the graph
def make_graph(data, title):
    plt.figure(figsize=(12,6))
    plt.plot(data.index, data['Close'], color='blue', linewidth=2, label="Close Price")
    plt.title(title, fontsize=14)
    plt.xlabel("Date")
    plt.ylabel("Closing Price (USD)")
    plt.legend()
    plt.grid(True)
    plt.show()

# Call the function for Tesla stock
make_graph(tesla_data, "Tesla Stock Price Over Time")
````
